```python
# import seaborn and othe required libs
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

$sin(flip) = sin(x+i*0.5)*(7-i)*flip$\,

$where\ i\in [0,7)$

$and\ x\in [0,14,100)$

```python
def sinplot(flip = 1):
    '''
    It plots sine curve with arguments-
    flip: type integer and default value = 1
    '''
    x = np.linspace(0, 14, 100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sinplot(1)
```

```python
sinplot(-1)
```

```python
# specify style in set style method of seaborn
# observe the changes in x and y axis for ticks
# And direction of ticks
# Experiment with this code to explore more
sns.set_style('ticks', 
              {'axes.grid': True, 'xtick.direction': 'in'})
sinplot()

sns.despine(left = False, bottom= False)
```

```python
# specify style in set style method of seaborn
sns.set_style('ticks', 
              {'axes.grid': True, 'xtick.direction': 'in'})
sinplot()

sns.despine(left = True, bottom= False)
```
