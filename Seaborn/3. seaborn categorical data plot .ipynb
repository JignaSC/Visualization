#### 1. ```catplot()```

**Full signature of catplot-**

```python
seaborn.catplot(x=None, y=None, hue=None, data=None, row=None, col=None, col_wrap=None, estimator=<function mean at 0x105c7d9e0>, ci=95, n_boot=1000, units=None, seed=None, order=None, hue_order=None, row_order=None, col_order=None, kind='strip', height=5, aspect=1, orient=None, color=None, palette=None, legend=True, legend_out=True, sharex=True, sharey=True, margin_titles=False, facet_kws=None, **kwargs)
```

For detailed info refer docs here - [catplot](https://seaborn.pydata.org/generated/seaborn.catplot.html)

````python
# lets apply catplot on tips data for column day vs total bill
sns.catplot(x = 'day', y = 'total_bill', data = tips)
````

````python
# change the x and y axis for horizontall plot
sns.catplot(y = 'day', x = 'total_bill', data = tips)
````

````python
# use jitter = False to avoid scattering as shown before
sns.catplot(x = 'day', y = 'total_bill', data = tips, jitter = False)
````

````python
# lets change kind = 'swarm' for swarm plot
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm')
````

````python
# lets add hue for size. Compare with above graph 
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'size')
````

````python
# We can also change the order of the graph plot
# by passing the user defined ordered list to order argument
# Compare with above graph 
sns.catplot(x = 'day', 
            y = 'total_bill', 
            data = tips, kind = 'swarm', 
            hue = 'size', 
            order=['Sun', 'Sat', 'Fri', 'Thur'])
````

````python
tips.head()
````

````python
# another example of ordered list over tips dataset
sns.catplot(x = 'smoker', y = 'tip', data = tips, order= ['No', 'Yes'])
````

````python
# lets change the kind to box.
# this will plot a box plot for us
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips)
````

````python
# let add hue = sex column in the above box plot
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'sex')
````

````python
# lets use dodge = False keyword -
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'sex', dodge = False)
````

````python
# lets load another dataset called diamonds dataset
diamonds = sns.load_dataset('diamonds')
diamonds.head()
````

````python
# lets see other kind which is boxen plot
# before plotting we have sorted diamonds dataset by color
sns.catplot(x = 'color', y = 'price', kind = 'boxen', data = diamonds.sort_values('color'))
````

````python
sns.catplot(x = 'color', y = 'price', kind = 'boxen', data = diamonds.sort_values('color'))
````

````python
sns.catplot(x = 'day', y = 'total_bill', kind = 'boxen', data = tips)
````

````python
# use kind = voilin for voilng plot
sns.catplot(x = 'total_bill', y = 'day', kind = 'violin', data = tips, split = True,)
````

````python
# use hue = sex
sns.catplot(x = 'total_bill', y = 'day', hue = 'sex', kind = 'violin', data = tips, split = True,)
````

#### 2. ```swarmplot()```

Full signature of swarmplot-

```python
seaborn.swarmplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor='gray', linewidth=0, ax=None, **kwargs)
```

For detailed info refer docs here - [swarmplot](https://seaborn.pydata.org/generated/seaborn.swarmplot.html)

````python
g = sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', inner = None, data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', color = 'k', size = 3, data = tips, ax = g.ax)
````
