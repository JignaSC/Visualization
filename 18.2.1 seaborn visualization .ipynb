Lets see the basic graphs availble and then we'll see them one by one in the subsequent notebooks

````python
# import seaborn
import seaborn as sns

# apply deafault style
sns.set()
````

````python
# load dataset using seaborn
iris = sns.load_dataset('iris')
````

````python
sns.scatterplot(x = iris.sepal_length, 
                y = iris.sepal_width);
````

````python

import matplotlib.pyplot as plt

# same plotting code as above in matplotlib!
plt.plot(iris.sepal_length, iris.sepal_width,'o')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width');
````

````python
# histogram in matplotlib
plt.hist(iris['sepal_length'], density=True, alpha=0.5);
````

````python
# seaborn alternative of above code with kde plot
sns.distplot(iris['sepal_length']);
````

````python
# kde plot only with shaded region
sns.kdeplot(iris['sepal_length'], shade=True);
````

````python
# kde plot only without shaded region
sns.kdeplot(iris['sepal_length'], shade=False);
````

````python
# import pandas
import pandas as pd

# get yeast dataset from the following URL
URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/yeast/yeast.data"
yeast_df = pd.read_table(URL,sep="\s+", header=None)
yeast_df.columns =['Sequence Name','mcg', 'gvh', 'alm', 'mit', 'erl', 'pox','vac', 'nuc',  'class_protein_localization' ]
yeast_df.head()
````

````python
cols = yeast_df.columns[1:-1]
for col in cols:
    plt.hist(yeast_df[col], density=True, alpha=0.5)
````

````python
sns.distplot(yeast_df.mcg)
````

````python
sns.distplot(yeast_df.alm)
````

````python
numerical_yeast_df = yeast_df.iloc[:, 1:-1]
numerical_yeast_df
````

````python
sns.kdeplot(x = 'mcg', y = 'alm', data = numerical_yeast_df[['mcg', 'alm']])
````

````python
with sns.axes_style('white'):
    sns.jointplot(yeast_df.mcg, yeast_df.alm, yeast_df, kind='kde');
````

````python
with sns.axes_style('white'):
    sns.jointplot(yeast_df.mcg, yeast_df.alm, yeast_df, kind='hex');
````

````python
yeast_df.info()
sns.pairplot(yeast_df, hue='class_protein_localization', height=2.5);
````
